```python
# mapping

yes_no_mapping = {'Yes' : 1 , 'No' : 0 }
```

```python
data['Partner'] = data['Partner'].map(yes_no_mapping) 
data['Dependents'] = data['Dependents'].map(yes_no_mapping) 
data['PhoneService'] = data['PhoneService'].map(yes_no_mapping)
```

```python
internetservicemapping = {'Yes':1 , 'No':0 , 'No internet service':0}

data['OnlineSecurity'] = data['OnlineSecurity'].map(internetservicemapping)
data['OnlineBackup'] = data['OnlineBackup'].map(internetservicemapping)
data['DeviceProtection'] = data['DeviceProtection'].map(internetservicemapping)
data['TechSupport'] = data['TechSupport'].map(internetservicemapping)
data['StreamingMovies'] = data['StreamingMovies'].map(internetservicemapping)

data['StreamingMovies'].value_counts()
```

```python
phoneservicemapping = {'Yes':1 , 'No':0 , 'No phone service':0}

data['MultipleLines'] = data['MultipleLines'].map(phoneservicemapping)
```

```python
dslfmapping = {'DSL' : 1 , 'Fiber optic' : 1 , 'No' : 0}
data['InternetService'] = data['InternetService'].map(dslfmapping)
```

```python
# Concatunate

data = pd.concat([data,sex,contract,paymethod], axis=1)
```

```python
data
```

```python
# Change 'True' and 'False' to 1 and 0

data['Male'] = data['Male'].astype(int)
data['OY'] = data['OY'].astype(int)
data['TY'] = data['TY'].astype(int)
data['Card'] = data['Card'].astype(int)
data['Check'] = data['Check'].astype(int)
```

```python
data
```

```python
# Remove Unnecessary columns

data.drop(['customerID','gender','PaymentMethod','Contract'], axis = 1, inplace=True)
```

```python
data
```

```python
# Encoding

from sklearn.preprocessing import LabelEncoder
lblEnc = LabelEncoder()
churn_data = data.copy()

for i in data.columns:
    churn_data[i] = lblEnc.fit_transform(data[i])
```

```python
churn_data
```

```python
X = churn_data.drop('Churn' , axis = 1)
Y = churn_data['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Divide data as train and test data

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(x_train.shape, '------> Training Feature Shape')
print(x_test.shape, '------> Testing Feature Shape')
print(y_train.shape, '------> Training Labels Shape')
print(y_test.shape, '------> Testing Labels Shape')
```

```python
# Import the algorithm

from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# Get predictions

predictions = lr.predict(x_test)
```

```python
predictions
```

# To get classification report

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test, predictions))
```

```python
import sklearn.metrics as sm
```

```python
accuracy = sm.accuracy_score(y_test,predictions)
```

```python
print('Accuracy Score = ', str(accuracy))
```
